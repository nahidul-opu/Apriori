```python
import pandas as pd
pd.set_option('display.max_colwidth',1)
```

```python
data = pd.read_csv("groceries.csv",error_bad_lines=False,header=None,warn_bad_lines=False)
```

```python
data.apply(data.value_counts).sum(axis=1).sort_values(ascending=False)
```

```python
records = []
for i in range((len(data))):
    records.append([str(data.values[i,j]) for j in range(0, data.shape[1])])
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder

te = TransactionEncoder()
te_data = te.fit(records).transform(records)
df = pd.DataFrame(te_data, columns = te.columns_)
df = df.drop(("nan"),axis=1)
df.head()
```

```python
from mlxtend.frequent_patterns import apriori

df1 = apriori(df, min_support=0.003, use_colnames = True, verbose=1)
df1.head()
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(df1, metric = "confidence", min_threshold = 0.5)
rules
```

## Own Dataset

```python
df=pd.read_csv("MLData.csv")
```

```python
columns=["Timestamp","Level","Class","Scale","Gender","Age","Residence","RelationshipStatus","FinanceState","CopeWithInstitute","RelationWithFamily","Pressure","AcademicResult","LivingPlace","SupportedBy","SocialMediaIn6","InferiorityComplex","MealSatisfaction","Health","OtherPositiveActivity","SleepTime"]
```

```python
df.columns=columns
```

```python
df=df.drop(['Scale','Timestamp'],axis=1)
def NameCopeWithInstitute(x):
    if x==1:
        return "DefinitelyNotSuccessful"
    elif x==2:
        return "ProbablyNotSuccessful"
    elif x==3:
        return "PossiblySuccessful"
    elif x==4:
        return "ProbablySuccessful"
    elif x==5:
        return "DefinitelySuccessful"
    

df['CopeWithInstitute']=df['CopeWithInstitute'].apply(NameCopeWithInstitute)
df['FinanceState']=df['FinanceState'].apply(lambda x: 'GoodFinance' if x=="Yes" else "BadFinance")
df['RelationWithFamily']=df['RelationWithFamily'].apply(lambda x: x+"RelationWithFamily")
df['Pressure']=df['Pressure'].apply(lambda x: x+"Pressure")
df['AcademicResult']=df['AcademicResult'].apply(lambda x: "Result"+x)
df['LivingPlace']=df['LivingPlace'].apply(lambda x: "LivingPlace"+x)
df['SupportedBy']=df['SupportedBy'].apply(lambda x: "SupportedBy"+x)
df['SocialMediaIn6']=df['SocialMediaIn6'].apply(lambda x: x+"SocialMediaIn6")
df['InferiorityComplex']=df['InferiorityComplex'].apply(lambda x: x+"InferiorityComplex")
df['MealSatisfaction']=df['MealSatisfaction'].apply(lambda x: x+"MealSatisfaction")
df['Health']=df['Health'].apply(lambda x: x+"Health")
df['OtherPositiveActivity']=df['OtherPositiveActivity'].apply(lambda x: x+"OtherPositiveActivity")
df['SleepTime']=df['SleepTime'].apply(lambda x: "SleepTime"+str(x))
```

```python
records = []
for i in range((len(df))):
    records.append([str(df.values[i,j]) for j in range(0, df.shape[1])])
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder

te = TransactionEncoder()
te_data = te.fit(records).transform(records)
df = pd.DataFrame(te_data, columns = te.columns_)
#df = df.drop(("nan"),axis=1)
df.head()
```

```python
from mlxtend.frequent_patterns import apriori

df1 = apriori(df, min_support=0.8, use_colnames = True, verbose=1)
df1.head()
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(df1, metric = "confidence", min_threshold = 0.5)
rules
```
