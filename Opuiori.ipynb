## Own Dataset

```python
df=pd.read_csv("MLData.csv")
columns=["Timestamp","Level","Class","Scale","Gender","Age","Residence","RelationshipStatus","FinanceState","CopeWithInstitute","RelationWithFamily","Pressure","AcademicResult","LivingPlace","SupportedBy","SocialMediaIn6","InferiorityComplex","MealSatisfaction","Health","OtherPositiveActivity","SleepTime"]
df.columns=columns
df=df.drop(['Scale','Timestamp'],axis=1)
def NameCopeWithInstitute(x):
    if x==1:
        return "Definitely Not Successful"
    elif x==2:
        return "Probably Not Successful"
    elif x==3:
        return "Possibly Successful"
    elif x==4:
        return "Probably Successful"
    elif x==5:
        return "Definitely Successful"
    

df['CopeWithInstitute']=df['CopeWithInstitute'].apply(NameCopeWithInstitute)
df['FinanceState']=df['FinanceState'].apply(lambda x: 'Good Finance' if x=="Yes" else "Bad Finance")
df['RelationWithFamily']=df['RelationWithFamily'].apply(lambda x: x+" Relation With Family")
df['Pressure']=df['Pressure'].apply(lambda x: x+" Pressure")
df['AcademicResult']=df['AcademicResult'].apply(lambda x: "Result "+x)
df['LivingPlace']=df['LivingPlace'].apply(lambda x: "Living Place "+x)
df['SupportedBy']=df['SupportedBy'].apply(lambda x: "Supported By "+x)
df['SocialMediaIn6']=df['SocialMediaIn6'].apply(lambda x: x+" SocialMediaIn6")
df['InferiorityComplex']=df['InferiorityComplex'].apply(lambda x: x+" InferiorityComplex")
df['MealSatisfaction']=df['MealSatisfaction'].apply(lambda x: x+" MealSatisfaction")
df['Health']=df['Health'].apply(lambda x: x+" HealthProblem")
df['OtherPositiveActivity']=df['OtherPositiveActivity'].apply(lambda x: x+" OtherPositiveActivity")
df['SleepTime']=df['SleepTime'].apply(lambda x: "SleepTime "+str(x))
records = []
for i in range((len(df))):
    records.append([str(df.values[i,j]) for j in range(0, df.shape[1])])
```

```python
frequent_itemsets,rules=Opuiori.apriori(records,min_sup=0.8,min_conf=0.5)
```
